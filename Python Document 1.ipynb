### Lists

##### Declaring Lists

```python
list_1 = []

names =('Max', 'Cindy', 'Faith', 'bob','Nate' )
print(names)
```

##### Indexing a List

```python
#list name(start, stop, step)

#single index
print(names[0])

#Print starting at index 1 going to end
print(names[1:])

#Print starting at the beginning of a list up until a number
print(names[:2])

#Print starting at index 1 and going up by 2 in each iteration
print(names[1::2])

#Print starting at the back and present in reverse order
print(names[::-1])
```

##### .append()

```python
names.append('Brittany')
print(names)
```

##### .insert()

```python
names.insert(1, 'Darrell')
print(names)
```

##### .pop()

```python
#Default to the last value if to parameter is given 
#Pop returns the element that was removed in case you want to assign it to a variable

my_name = names.pop[2]
print(my_name)
print(names)
```


##### .remove()

```python
#Value to be removed rather than index
# names.remove('bob')
# print(names)

#remove multiple brittanys from list
while 'Brittany' in names:
    names.remove('Brittany')
    print(names)
```

##### del()

```python
#Goes by index rather than value
#Be careful with del. it can cause index errors if not used carefully

del(names[1])
print(names)
```

##### Concatenating Two Lists

```python
#Will append two lists together, will NOT add the values!

list_2 = [0, 1, 2]
list_3 = [3, 4, 5]

large_list = list_2 + list_3
print(large_list)
```

##### Lists Within Lists

```python
#Lists can hold any other type of element, including other lists
#They can go as deep as you want. This is called nested lists

names = ['Max', 'Sam', 'Josh',['Sally',['Sue', 'Jim'], 'Tamika']]
print(names)
print(names[3][1][1])
```

##### Looping Through Lists

```python
#Two ways to loop through lists: One is by index; other by using 'in' keyword

#By index
for i in range(len(names)):
    print(names[i])

#loop with 'in'
for i in names:
    print(i)
```

# Exercise 3 <br>
<p>Take in a users input for their age, if they are younger than 18 print kids, if they're 18 to 65 print adults, else print seniors</p>
